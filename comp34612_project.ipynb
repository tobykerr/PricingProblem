**Leader’s Specifications**

Assume that the leader’s strategy (i.e., price) is $u_L$ and the follower’s strategy (i.e., price) is $u_F$. The specifics regarding the leader’s strategy and profits are as follows:
*   **Unit cost**: $c_L=£1.00$
*   **Strategy space**:
      1.   When playing with MK1, MK2, MK4 and MK5, the leader’s strategy space is $U_L = [1.00, +∞)$

*   **Demand Model**: (i.e., price-sale relationship): $S_L(u_L, u_F)=100 - 5u_L + 3u_F$    
*   **Daily Profit**: $(u_L - c_L)S_L(u_L, u_F)$
*   **Objective**: The leader’s objective of playing is to maximise the accumulated profit for the next 30 days.
*   **Assumption**: For simplicity, it is assumed that the leader’s unit cost, strategy space, and the demand model remain unchanged throughout the entire period of 130 days.

## **Game Rules and Playing Scenarios**

**Game Rules**

On day $t$, the leader announces the price, $u_L (t)$, first. After learning the leader’s price, the follower will choose the responding price, $u_F (t)$. This process occurs over a period of $t=1,2,…,130$, where $t=1,2,…,100$ corresponds to the days for which historical data is available, and $t=101,102,…,130$ are the days during which the game will be played repeatedly.


**Game Playing Scenarios**

In this setup, your program, acting as the leader, will engage in three separate games against three different competitors, identified as followers MK1, MK2, and MK3.

For each of the repeated game, the leader selects his learning method to learn the follower’s reaction function based on the provided set of historical data which spans from $t=1,2,…,100$. During the next 30 days ($t=101,102,…,130$), the leader announces a price, $u_L (t)$, to the game platform daily. After the leader's price is announced, the follower will respond with the price, $u_F (t)$. Upon receiving the follower's price, $u_F (t)$, from the game platform, the leader may use this new information to refine their knowledge/ update the learned reaction function and decide his price of the next day. In the other words, the leader (i.e., your program) should be able to take the follower’s price of the previous day from the game platform for updating, and then decide and send his price of the next day to the game platform.

It should be emphasised that the follower’s strategy and payoff function is subject to the changing and time varying environment. That is, the parameters in the follower’s payoff function may not be the same every day.

## Base Leader

This is the base class for all leader subclasses.

```python
class Leader:
    _subclass_registry = {}

    def __init__(self, name, engine):
        self.name = name
        self.engine = engine

    @classmethod
    def cleanup_old_subclasses(cls):
        """
        A function to remove old subclasses before defining new ones.
        """
        existing_subclasses = list(cls.__subclasses__())

        for subclass in existing_subclasses:
            subclass_name = subclass.__name__
            if subclass_name in cls._subclass_registry:
                del cls._subclass_registry[subclass_name]
                del subclass
        gc.collect()

    @classmethod
    def update_subclass_registry(cls):
        """
        A function to update registry after cleaning up old subclasses.
        """
        cls.cleanup_old_subclasses()
        cls._subclass_registry = {subclass.__name__: subclass for subclass in cls.__subclasses__()}

    def new_price(self, date):
        """
        A function for setting the new price of each day.
        :param date: date of the day to be updated
        :return: (float) price for the day
        """
        pass

    def get_price_from_date(self, date):
        """
        A function for getting the price set on a date.
        :param date: (int) date to get the price from
        :return: a tuple (leader_price, follower_price)
        """
        return self.engine.exposed_get_price(date)


    def start_simulation(self):
        """
        A function runs at the beginning of the simulation.
        """
        pass

    def end_simulation(self):
        """
        A function runs at the beginning of the simulation.
        """
        pass
```

```python
# ============================================================
# Shared helper functions
# ============================================================

def _safe_get_price(leader_obj, date):
    """
    Safely read (leader_price, follower_price) for a given date.
    Returns None if unavailable.
    """
    try:
        lp, fp = leader_obj.get_price_from_date(date)
        return float(lp), float(fp)
    except Exception:
        return None


def _load_initial_history(leader_obj):
    """
    Load the initial historical data (days 1..history_days).
    """
    if leader_obj.loaded_history:
        return

    leader_obj.X = []
    leader_obj.y = []
    leader_obj.seen_dates = set()

    for d in range(1, leader_obj.history_days + 1):
        obs = _safe_get_price(leader_obj, d)
        if obs is not None:
            lp, fp = obs
            leader_obj.X.append(lp)
            leader_obj.y.append(fp)
            leader_obj.seen_dates.add(d)

    leader_obj.loaded_history = True


def _ingest_previous_day_if_available(leader_obj, date):
    """
    At live date t, try to ingest data from day t-1 if not already added.
    """
    prev_date = date - 1
    if prev_date <= leader_obj.history_days:
        return
    if prev_date in leader_obj.seen_dates:
        return

    obs = _safe_get_price(leader_obj, prev_date)
    if obs is not None:
        lp, fp = obs
        leader_obj.X.append(lp)
        leader_obj.y.append(fp)
        leader_obj.seen_dates.add(prev_date)


def _fit_ols_params(X_vals, y_vals):
    """
    Ordinary least squares fit for:
        y = a + b x
    Returns (a, b).
    """
    n = len(X_vals)

    if n == 0:
        return 0.0, 1.0
    if n == 1:
        return float(y_vals[0]), 0.0

    X_arr = np.asarray(X_vals, dtype=float)
    y_arr = np.asarray(y_vals, dtype=float)

    A = np.column_stack([np.ones_like(X_arr), X_arr])
    coeffs, _, _, _ = np.linalg.lstsq(A, y_arr, rcond=None)

    a = float(coeffs[0])
    b = float(coeffs[1])
    return a, b


def _predict_follower_price(a, b, leader_price):
    return a + b * leader_price


def _expected_profit(cost, a, b, leader_price):
    follower_hat = _predict_follower_price(a, b, leader_price)
    demand = 100.0 - 5.0 * leader_price + 3.0 * follower_hat
    demand = max(0.0, demand)  # conservative clamp
    return (leader_price - cost) * demand


def _optimise_price(cost, a, b, min_price, max_price, grid_size):
    """
    Grid-search for the best leader price.
    """
    prices = np.linspace(min_price, max_price, grid_size)

    best_price = float(min_price)
    best_profit = -float("inf")

    for p in prices:
        prof = _expected_profit(cost, a, b, float(p))
        if prof > best_profit:
            best_profit = prof
            best_price = float(p)

    return best_price


def _rls_initial_state(P0_scale=1000.0):
    """
    Initialise recursive least squares state.
    theta = [a, b]^T
    """
    theta = np.zeros((2, 1), dtype=float)
    P = float(P0_scale) * np.eye(2, dtype=float)
    return theta, P


def _rls_update(theta, P, x, y, forgetting_factor):
    """
    One recursive least squares update for model:
        y = a + b x
    where theta = [a, b]^T
    """
    phi = np.array([[1.0], [float(x)]], dtype=float)
    y = float(y)
    lam = float(forgetting_factor)

    denom = lam + float(phi.T @ P @ phi)
    K = (P @ phi) / denom

    pred = float(phi.T @ theta)
    err = y - pred

    theta_new = theta + K * err
    P_new = (P - K @ phi.T @ P) / lam

    a = float(theta_new[0, 0])
    b = float(theta_new[1, 0])

    return theta_new, P_new, a, b


def _initialise_common_state(
    leader_obj,
    min_price,
    max_price,
    grid_size=500,
    history_days=100,
    cost=1.0,
):
    """
    Common attribute setup for all leader classes.
    """
    leader_obj.min_price = float(min_price)
    leader_obj.max_price = float(max_price)
    leader_obj.grid_size = int(grid_size)
    leader_obj.history_days = int(history_days)
    leader_obj.cost = float(cost)

    leader_obj.X = []
    leader_obj.y = []
    leader_obj.seen_dates = set()
    leader_obj.loaded_history = False

    # Current linear model: follower_price = a + b * leader_price
    leader_obj.a = 0.0
    leader_obj.b = 1.0


# ============================================================
# Offline OLS baselines
# ============================================================

class MK1_OfflineOLSLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=30.0, grid_size=500)

    def start_simulation(self):
        _load_initial_history(self)
        self.a, self.b = _fit_ols_params(self.X, self.y)

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


class MK2_OfflineOLSLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=30.0, grid_size=500)

    def start_simulation(self):
        _load_initial_history(self)
        self.a, self.b = _fit_ols_params(self.X, self.y)

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


class MK3_OfflineOLSLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=15.0, grid_size=500)

    def start_simulation(self):
        _load_initial_history(self)
        self.a, self.b = _fit_ols_params(self.X, self.y)

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


# ============================================================
# Online RLS baselines
# ============================================================

class MK1_RLSLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=30.0, grid_size=500)
        self.forgetting_factor = 0.98
        self.P0_scale = 1000.0
        self.theta, self.P = _rls_initial_state(self.P0_scale)

    def start_simulation(self):
        _load_initial_history(self)
        self.theta, self.P = _rls_initial_state(self.P0_scale)

        for x, y in zip(self.X, self.y):
            self.theta, self.P, self.a, self.b = _rls_update(
                self.theta, self.P, x, y, self.forgetting_factor
            )

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        prev_date = date - 1
        if prev_date > self.history_days and prev_date not in self.seen_dates:
            obs = _safe_get_price(self, prev_date)
            if obs is not None:
                x, y = obs
                self.X.append(x)
                self.y.append(y)
                self.seen_dates.add(prev_date)
                self.theta, self.P, self.a, self.b = _rls_update(
                    self.theta, self.P, x, y, self.forgetting_factor
                )

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


class MK2_RLSLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=30.0, grid_size=500)
        self.forgetting_factor = 0.98
        self.P0_scale = 1000.0
        self.theta, self.P = _rls_initial_state(self.P0_scale)

    def start_simulation(self):
        _load_initial_history(self)
        self.theta, self.P = _rls_initial_state(self.P0_scale)

        for x, y in zip(self.X, self.y):
            self.theta, self.P, self.a, self.b = _rls_update(
                self.theta, self.P, x, y, self.forgetting_factor
            )

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        prev_date = date - 1
        if prev_date > self.history_days and prev_date not in self.seen_dates:
            obs = _safe_get_price(self, prev_date)
            if obs is not None:
                x, y = obs
                self.X.append(x)
                self.y.append(y)
                self.seen_dates.add(prev_date)
                self.theta, self.P, self.a, self.b = _rls_update(
                    self.theta, self.P, x, y, self.forgetting_factor
                )

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


class MK3_RLSLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=15.0, grid_size=500)
        self.forgetting_factor = 0.97
        self.P0_scale = 1000.0
        self.theta, self.P = _rls_initial_state(self.P0_scale)

    def start_simulation(self):
        _load_initial_history(self)
        self.theta, self.P = _rls_initial_state(self.P0_scale)

        for x, y in zip(self.X, self.y):
            self.theta, self.P, self.a, self.b = _rls_update(
                self.theta, self.P, x, y, self.forgetting_factor
            )

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        prev_date = date - 1
        if prev_date > self.history_days and prev_date not in self.seen_dates:
            obs = _safe_get_price(self, prev_date)
            if obs is not None:
                x, y = obs
                self.X.append(x)
                self.y.append(y)
                self.seen_dates.add(prev_date)
                self.theta, self.P, self.a, self.b = _rls_update(
                    self.theta, self.P, x, y, self.forgetting_factor
                )

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


# ============================================================
# Online moving-window OLS baselines
# ============================================================

class MK1_MovingWindowLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=30.0, grid_size=500)
        self.window_size = 30

    def _refit_window(self):
        if len(self.X) <= self.window_size:
            self.a, self.b = _fit_ols_params(self.X, self.y)
        else:
            self.a, self.b = _fit_ols_params(self.X[-self.window_size:], self.y[-self.window_size:])

    def start_simulation(self):
        _load_initial_history(self)
        self._refit_window()

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        _ingest_previous_day_if_available(self, date)
        self._refit_window()

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


class MK2_MovingWindowLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=30.0, grid_size=500)
        self.window_size = 30

    def _refit_window(self):
        if len(self.X) <= self.window_size:
            self.a, self.b = _fit_ols_params(self.X, self.y)
        else:
            self.a, self.b = _fit_ols_params(self.X[-self.window_size:], self.y[-self.window_size:])

    def start_simulation(self):
        _load_initial_history(self)
        self._refit_window()

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        _ingest_previous_day_if_available(self, date)
        self._refit_window()

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass


class MK3_MovingWindowLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        _initialise_common_state(self, min_price=1.0, max_price=15.0, grid_size=500)
        self.window_size = 25

    def _refit_window(self):
        if len(self.X) <= self.window_size:
            self.a, self.b = _fit_ols_params(self.X, self.y)
        else:
            self.a, self.b = _fit_ols_params(self.X[-self.window_size:], self.y[-self.window_size:])

    def start_simulation(self):
        _load_initial_history(self)
        self._refit_window()

    def new_price(self, date):
        if not self.loaded_history:
            self.start_simulation()

        _ingest_previous_day_if_available(self, date)
        self._refit_window()

        return _optimise_price(
            cost=self.cost,
            a=self.a,
            b=self.b,
            min_price=self.min_price,
            max_price=self.max_price,
            grid_size=self.grid_size,
        )

    def end_simulation(self):
        pass

class TreeSearchLeader(Leader):
    def __init__(self, name, engine):
        super().__init__(name, engine)
        self.cost = 1.0
        self.history_days = 100
        self.simulation_runs = 1000  # Number of 'futures' to simulate per price

        # Model parameters (Learned once from history)
        self.a = 0.0
        self.b = 0.0
        self.noise_std = 0.1
        self.is_constrained = False

    def start_simulation(self):
        """Analyze history once to build the 'World Model'."""
        l_prices = []
        f_prices = []

        for d in range(1, self.history_days + 1):
            obs = self.get_price_from_date(d)
            if obs:
                l_prices.append(obs[0])
                f_prices.append(obs[1])

        # Detect if we are playing MK3 or MK6 (Price cap of 15.0)
        if max(l_prices) <= 15.1:
            self.is_constrained = True

        # 1. Calculate the Linear Model (u_F = a + b * u_L)
        # This is 'Offline Learning' - calculated once before the 30 days start.
        A = np.vstack([np.ones(len(l_prices)), l_prices]).T
        self.a, self.b = np.linalg.lstsq(A, f_prices, rcond=None)[0]

        # 2. Calculate historical noise (standard deviation of residuals)
        # This allows the 'Tree' to explore uncertainty.
        preds = self.a + self.b * np.array(l_prices)
        self.noise_std = np.std(np.array(f_prices) - preds)

    def new_price(self, date):
        """Simulation-based optimization for the remaining days."""
        days_left = 131 - date

        # Define candidate prices (Grid search for the best root move)
        upper_limit = 15.0 if self.is_constrained else 25.0
        candidates = np.linspace(1.1, upper_limit, 50)

        best_price = 1.1
        max_expected_total_profit = -float('inf')

        for p_test in candidates:
            total_profit_from_sims = []

            for _ in range(self.simulation_runs):
                # Rollout: Simulate the sum of profit for the remaining window
                sim_accumulated = 0
                for _t in range(days_left):
                    # Add historical noise to the follower's expected response
                    # This makes the strategy robust for MK4-MK6
                    noise = np.random.normal(0, self.noise_std)
                    u_f_sim = self.a + self.b * p_test + noise

                    # Demand: 100 - 5*uL + 3*uF
                    demand = 100 - 5 * p_test + 3 * u_f_sim
                    profit = (p_test - self.cost) * max(0, demand)
                    sim_accumulated += profit

                total_profit_from_sims.append(sim_accumulated)

            # Use the mean result of simulations to find the 'safest' optimal price
            expected_val = np.mean(total_profit_from_sims)
            if expected_val > max_expected_total_profit:
                max_expected_total_profit = expected_val
                best_price = p_test

        return float(best_price)

    def end_simulation(self):
        pass
```
